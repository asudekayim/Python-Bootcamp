Bir şirketin **8 çalışana ait 6 aylık performans skorları** NumPy ile analiz edilecektir.

- Satırlar → Çalışanlar
- Sütunlar → Aylar
- Değerler → Performans puanı (0–100)

```python
import numpy as np
```

```python
performans = np.random.randint(50,100, size=(8,6))
```

```python
performans
```

```python
performans.shape
```

```python
performans.ndim
```

```python
performans.size
```

```python
# ilk çalışanın tüm ayları
performans[0,:]
```

```python
# son ayın tüm çalışanları
performans[:,5]
```

```python
# ilk 3 çalışanın ilk 2 ayını seç
performans[:3,:2]
```

```python
# belirli bir çalışanın belirli ay skorunu al
performans[4,2]
```

```python
# sub-matrix
subMatrix = performans[2:5,1:5]
```

```python
subMatrix
```

```python
yeniPerformans = performans + 5
```

```python
yeniPerformans
```

```python
normalize = (performans / performans.max()) * 100
```

```python
normalize
```

```python
ay1 = performans[:,0]
ay2 = performans[:,1]
```

```python
print(ay1 > ay2)
```

```python
calisanBazli = performans.mean(axis=1)
```

```python
calisanBazli
```

```python
ayBazli = performans.mean(axis=0)
```

```python
ayBazli
```

```python
# axis = 0 aşağı doğru, axis = 1 yana doğru çalışır
```

```python
yuksekPerformans = performans.argmax()
dusukPerformans = performans.argmin()
```

```python
dusukPerformans
```

```python
performans.reshape(12,4)
```

```python
performans.ravel()
```

```python
print(f"En başarılı çalışan: {performans.argmax()}")
print(f"En tembel çalışan: {performans.argmin()}")
print(f"En yüksek ay ortalaması: {ayBazli.max()}")
print(f"En yüksek çalışan ortalaması: {calisanBazli.max()}")
print(f"Genel ortalama performans: {performans.mean()}")
```
