```python
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

```python
# Toplam sipariş sayısı
```

```python
orderSum = len(orders)
```

```python
orderSum
```

```python
# Ödenmiş ve ödenmemiş sipariş sayısı
```

```python
paidSum = 0
notPaidSum = 0

if orders[0]["is_paid"] == True:
    paidSum += 1
else:
    notPaidSum += 1
    
if orders[1]["is_paid"] == True:
    paidSum += 1
else:
    notPaidSum += 1
    
if orders[2]["is_paid"] == True:
    paidSum += 1
else:
    notPaidSum += 1

if orders[3]["is_paid"] == True:
    paidSum += 1
else:
    notPaidSum += 1
```

```python
paidSum
```

```python
notPaidSum
```

```python
# Müşteri  Listesi
```

```python
customerList = [orders[0]["customer"],orders[1]["customer"],orders[2]["customer"],orders[3]["customer"]]
```

```python
customerList
```

```python
# Tekrar eden müşteri temizlemesi
```

```python
customerSet = set(customerList)
```

```python
customerSet
```

```python
# Tüm siparişlerdeki ürünler
```

```python
productList = orders[0]["items"] + orders[1]["items"] + orders[2]["items"] + orders[3]["items"]
```

```python
productList
```

```python
# aynı ürün tekrarı
```

```python
if productList.count("Laptop") > 1:
    print(f"Laptop {productList.count("Laptop")} tane var.")
else:
    print(f"Laptop {productList.count("Laptop")} tane var.")
    
if productList.count("Mouse") > 1:
    print(f"Mouse {productList.count("Mouse")} tane var.")
else:
    print(f"Mouse {productList.count("Mouse")} tane var.")
    
if productList.count("Keyboard") > 1:
    print(f"Keyboard {productList.count("Keyboard")} tane var.")
else:
    print(f"Keyboard {productList.count("Keyboard")} tane var.")
    
if productList.count("Monitor") > 1:
    print(f"Monitor {productList.count("Monitor")} tane var.")
else:
    print(f"Monitor {productList.count("Monitor")} tane var.")
    
if productList.count("HDMI Cable") > 1:
    print(f"HDMI Cable {productList.count("HDMI Cable")} tane var.")
else:
    print(f"HDMI Cable {productList.count("HDMI Cable")} tane var.")
    
```

```python
# unique ürünleri set ile ayırma
```

```python
productSet = set(productList)
```

```python
productSet
```
