```python
import random

scores = []

while len(scores) <20:
    score = random.randint(0,6)
    # ekleme yaptım
    print(score)

    if score ==0:
        break

    if score ==6:
        continue

    scores.append(score)
```

```python
# Kaç geçerli puan toplandığını yazdır
```

```python
print(scores)
```

```python
isValid = len(scores)
```

```python
isValid
```

```python
# Burada while döngünün en fazla 20 kere çalışması için yazılmıştır. 20 tekrarda da bitebilir ama daha az bir tekrarda da bitebilir. 
# Belirsizlik olduğu için while kullanılmıştır.
```

```python
# Ortalama hesaplama
```

```python
sum = 0
for score in scores:
    sum += score
print(sum)
```

```python
mean = sum / len(scores)
```

```python
mean
```

```python
# round()
```

```python
print(round(mean, 2))
```

```python
minScore
```

```python
maxScore = max(scores)
```

```python
maxScore
```

```python
# Memnuniyet Analizi ve Erken Uyarı 
```

```python
ones = 0
memnun = 0
kararsiz = 0
memnunDegil = 0

for score in scores:
    if score >= 4:
        print("Memnun")
        memnun += 1
    elif score == 3:
        print("Kararsız")
        kararsiz += 1
    else:
        print("Memnun değil")
        memnunDegil += 1

    if score == 1:
        ones += 1
        if ones == 3:
            print("⚠️ Kritik memnuniyetsizlik tespit edildi")
            break
        else:
            ones = 0
```
