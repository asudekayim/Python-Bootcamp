# OOP Neden Gerekli?
OOP, nesne tabanlı programlama demektir. Projelerde class denen yapıları kullanarak nesneleri oluşturur ve özelliklerinin kullanılmasını sağlar. Kod tekrarını azaltır çünkü inheritance(miras) ile özellikleri alabilir.
# Kullanılan OOP Kavramları
- class
- inheritance
- methods

```python
# Book sınıfı
```

```python
class Book():
    def __init__(self, title, author, year, is_available = True):
        self.title = title
        self.author = author
        self.year = year
        self.is_available = is_available

    def get_info(self):
        print(f"Başlık: {self.title} \nYazar: {self.author} \nYear: {self.year} \nMüsait mi: {self.is_available}")
```

```python
kitap = Book("Çalıkuşu","Reşat Nuri Güntekin",1922,True)
```

```python
kitap.get_info()
```

```python
# Member sınıfı
```

```python
class Member():
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id

    def borrow_book(self,book):
        if book.is_available:
            print(f"Kitap müsait. {self.member_id} nolu {self.name} tarafından alındı.")
            book.is_available = False
        else:
            print(f"Kitap müsait değil! {self.member_id} nolu {self.name} tarafından alınmış.")
        
    def return_book(self,book):
        book.is_available = True
        print(f"Kitap {self.member_id} nolu {self.name} tarafından iade edildi.")
```

```python
member = Member("Asude",100)
```

```python
kitap = Book("Çalıkuşu","Reşat Nuri Güntekin",1922,True)
```

```python
kitap.get_info()
```

```python
member.borrow_book(kitap)
```

```python
kitap.get_info()
```

```python
member.borrow_book(kitap)
```

```python
member.return_book(kitap)
```

```python
kitap.get_info()
```

```python
# inheritance-digitalbook
```

```python
class DigitalBook(Book):
    def __init__(self, file_size, title, author, year):
        Book.__init__(self, title, author, year)
        self.file_size = file_size

    def download(self):
        print(f"{self.file_size} MB dosya indiriliyor...")
```

```python
kitap2 = DigitalBook(54, "Hayvan Çiftliği", "George Orwell", 1945)
```

```python
kitap2.download()
```

```python
kitap2.get_info()
```
