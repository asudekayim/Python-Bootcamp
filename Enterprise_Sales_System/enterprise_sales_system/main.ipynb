```python
from data_generator import DataGenerator 
import preprocessing as pre
import analytics as anl
from reporting import Report
```

```python
satis = DataGenerator(100,42)
```

```python
pre.missing_values(df)
```

```python
df_final = pre.add_features(df_filled)
```

```python
product_perf
```

```python
monthly_perf = anl.get_monthly_performance(df_final)
```

```python
monthly_perf
```

```python
best_sellers
```

```python
df_dict = {
    "Şehir Analizi":   city_perf,
    "Ürün Analizi":    product_perf,
    "Aylık Özet":      monthly_perf,
    "En Çok Satanlar": best_sellers
}
```

```python
print(f"Toplam Satış     : {toplam_satis:,.0f} TL")
print(f"En Kârlı Şehir   : {en_karli_sehir}")
print(f"En İyi Ay        : {en_iyi_ay}")
```
