```python
import pandas as pd
import numpy as np
```

```python
customers = {
    "CustomerID" : [1,2,3,4],
    "CustomerName" : ["Furkan","Berkay","Beyza","Selin"],
    "City" : ["Samsun","Ankara","Denizli","Muğla"],
    "Age" : [23,np.nan,21,np.nan]
}
```

```python
customerDF = pd.DataFrame(customers)
```

```python
customerDF
```

```python
sales = {
    "SaleID" : [2,1,4,5],
    "CustomerID" : [1,2,3,4],
    "Product" : ["Bilgisayar","Telefon","Kulaklık","Bilgisayar"],
    "Amount" : [30000,25000,np.nan,np.nan]
}
```

```python
salesDF = pd.DataFrame(sales)
```

```python
salesDF
```

```python
regions = {
    "City" : ["Samsun","Denizli","Ankara","Muğla"],
    "Region" : ["Karadeniz","Ege","İç Anadolu","Ege"]
}
```

```python
regionsDF = pd.DataFrame(regions)
```

```python
regionsDF
```

```python
# eksik değerlerin tespiti
```

```python
customerDF.isnull()
```

```python
# Yaş için ortalama ile doldur
```

```python
customerDF["Age"] = customerDF["Age"].fillna(customerDF["Age"].mean())
```

```python
customerDF
```

```python
# satış tutarı eksik olan kayıtları sil
```

```python
salesDF
```

```python
salesDF = salesDF.dropna(subset=["Amount"])
```

```python
salesDF
```

```python
# merge
# customers + sales → CustomerID üzerinden birleştir
```

```python
mergedDF = pd.merge(customerDF,salesDF, on="CustomerID")
```

```python
mergedDF
```

```python
# regions ile birleştirme
```

```python
sonMergedDF = pd.merge(mergedDF, regionsDF)
```

```python
sonMergedDF
```

```python
# group by şehir bazlı
```

```python
groupSehir = sonMergedDF.groupby("City" ).agg(["sum"])
```

```python
groupSehir
```

```python
# group by bölge bazlı ort satış
```

```python
groupBolge = sonMergedDF.groupby("Region").agg({"Amount": "mean"})
```

```python
groupBolge
```

```python
# ürün bazlı satış sayısı
```

```python
groupUrun = sonMergedDF.groupby("Product").agg(["count"])
```

```python
groupUrun
```

```python
# concat
# farklı aylara ait iki satış tablosu oluşturma
```

```python
ocak = pd.DataFrame({
    "SaleID": [1, 2, 3],
    "Product": ["Bilgisayar", "Telefon", "Kulaklık"],
    "Amount": [30000, 25000, 1500],
    "Month": ["Ocak", "Ocak", "Ocak"]
})
```

```python
ocak
```

```python
subat = pd.DataFrame({
    "SaleID": [4, 5, 6],
    "Product": ["Bilgisayar", "Telefon", "Kulaklık"],
    "Amount": [32000, 27000, 2000],
    "Month": ["Şubat", "Şubat", "Şubat"]
})
```

```python
subat
```

```python
# satır bazlı birleştirme
```

```python
satislar = pd.concat([ocak, subat], axis=0)
```

```python
satislar
```

```python
# koşullu filtreleme
```

```python
booleanFrame = satislar["Amount"] > 20000
```

```python
satislar[booleanFrame]
```

```python
# yeni sütun oluşturma
```

```python
satislar["KDVli Satış"] = satislar["Amount"] * 1.5
```

```python
satislar
```

```python
# apply kullanımı
```

```python
print(satislar['Amount'].dtype)
```

```python
def new_amount(deger):
    return int(deger) + 2500
```

```python
satislar['Amount'] = satislar['Amount'].apply(new_amount)
```

```python
satislar
```

```python
# excel e yazma
```

```python
satislar.to_excel("final_sales_report.xlsx")
```

```python
# çıktılar
```

```python
# 'Amount' sütunundaki en büyük değerin olduğu satırı bul ve o satırın 'Product' ismini al
en_karli_urun = satislar.loc[satislar['Amount'].idxmax(), 'Product']
```

```python
print(f"Toplam Satış: {satislar["Amount"].sum()}")
print(f"En kârlı ürün: {en_karli_urun}")
```
