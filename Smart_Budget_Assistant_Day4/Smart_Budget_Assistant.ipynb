```python
# Harcama alma fonksiyonu
```

```python
expenses = []
def get_expenses():
    
    while True:
        try:
            harcama = int(input("Kaç harcama gireceksiniz?"))
            if harcama < 1:
                print("Lütfen geçerli bir sayı giriniz!!")
                continue
            else:
                print(f"Harcama sayısı: {harcama}")
                break 

        except:
            print("Geçerli bir sayı giriniz!")

    i = 0
    while i < harcama:
        try:
            price = float(input(f"TL cinsinden {i+1}. harcamayı giriniz: "))
            if price < 0:
                print("Geçersiz harcama tutarı!")
                continue
            else:
                expenses.append(price)
                i += 1

        except:
            print("Geçerli bir tutar giriniz!")
```

```python
get_expenses()
```

```python
expenses
```

```python
# Toplam ve ortalama alma fonksiyonu
```

```python
def calculate_summary(expenses):
    sumExpenses = 0
    sumExpenses = round(sum(expenses), 2)

    meanExpenses = round(sumExpenses/len(expenses), 2)
    return sumExpenses, meanExpenses
```

```python
calculate_summary(expenses)
```

```python
# *args ile esnek hesaplama
```

```python
def flexible_total(*args):
    summary = sum(args)
    argSum = len(args)
    print(f"Girilen değerlerin toplamı: {summary}")
    print(f"Girilen argüman sayısı: {argSum}")
```

```python
flexible_total(120)
```

```python
flexible_total(571,882)
```

```python
# En büyük/en küçük harcama
```

```python
def find_extremes():
    maxExpense = max(expenses)
    minExpense = min(expenses)
    return maxExpense, minExpense
```

```python
find_extremes()
```
